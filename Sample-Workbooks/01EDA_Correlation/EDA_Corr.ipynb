#### Loading packages and data


```r
library(ggplot2)
library(corrplot)
library(ggcorrplot)
```

```r
data_ratings <- read.csv("ratings.csv")
```

### Displaying Correlation

```r
cor(data_ratings$RATING, data_ratings$QUANTITY_DEMANDED)
```

```r
plot(data_ratings$RATING, data_ratings$QUANTITY_DEMANDED)
```

```r
cor.test(data_ratings$RATING, data_ratings$QUANTITY_DEMANDED)
```
